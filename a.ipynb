```python
# 2.
# Suppose a coin has probability p of falling heads up. If we flip the coin many times, we would expect
# the proportion of heads to be near p. Take p = 0.3 and n = 1000 and simulate n coin flips. Plot the
# proportion of heads as a function of n. Repeat for p = 0.03.
n = 1000
p = 0.3
x = np.arange(1, n + 1)
y = np.zeros(n)
heads_so_far = 0
for i in range(n):
    # flip a coin
    if np.random.random() < p:
        heads_so_far += 1
    # record the proportion of heads
    y[i] = heads_so_far / (i + 1)

# plot the proportion of heads as a function of n
plt.plot(x, y, 'r-', label='simulated')
# plot the expected proportion of heads
plt.plot(x, p * x / n, 'b-', label='expected')
plt.xlabel('n', color='white')
plt.ylabel('proportion of heads', color='white')
plt.xticks(color='white')
plt.yticks(color='white')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# 12. For the airline-passenger-satisfaction data set
# 1. Visualise the number of male to female passengers using a histogram.
# 2. Visualise the number of loyal customers to disloyal customer using a pie chart.
# 3. What is the average delay in departure experienced by male passengers?
# 4. What percentage of total passengers are below the age of 40?
# 5. What is the shortest and longest distance a flight travels?
# 6. Visualise the age of passengers using a box plot and explain your findings.
import pandas as pd
df = pd.read_csv('./airline_passenger_satisfaction_dataset.csv')
df
```

```python
# 1. 
genders = []
for i in range(len(df)):
    if df['Gender'][i] == 'Male':
        genders.append(1)
    else:
        genders.append(2)
plt.hist(genders, zorder=3)
plt.grid(True, zorder=0)
plt.xticks(np.arange(1, 3), ["Male", "Female"], color='white')
plt.yticks(color='white')
plt.show()
```

```python
# 2.
loyal = [0, 0]
for i in range(len(df)):
    if df['Customer Type'][i][0] == 'L':
        loyal[0] += 1
    else:
        loyal[1] += 1
plt.pie(loyal, labels=["Loyal", "Disloyal"], autopct='%1.1f%%', shadow=True, startangle=90)
# Set label color to white
plt.legend()
plt.show()
```

```python
# 3.
total_delay = 0
for i in range(len(df)):
    if df['Gender'][i] == 'Male':
        total_delay += df['Departure Delay in Minutes'][i]
print(f"Average Departure Delay Faced by Males: {total_delay/len(df[df['Gender'] == 'Male'].index)}")
```

```python
# 4.
percent = len(df[df['Age'] < 40].index)/len(df.index)
print(f"{percent*100}% Passengers are below 40")
```

```python
# 5.
print(f"Minimum Flight Distance: {min(df['Flight Distance'])}")
print(f"Maximum Flight Distance: {max(df['Flight Distance'])}")
```

```python
# 6.
B = plt.boxplot(df['Age'])
print([item.get_ydata() for item in B['whiskers']])
plt.yticks(color='white')
plt.grid(True)
plt.show()
# FINDINGS :
# The age distribution is skewed right.
# Median is 40
# Maximum age is 85
# Minimum age is 7
# 25% of the passengers are below 27
# 75% of the passengers are below 51
```

```python
# 10. For the Rdatasets data set
# 1. Visualise the gender distribution using a pie chart and calculate what percentage of teachers are male.
# 2. If one teacher is selected at random, what is the probability that the selected teacher is a professor?
# 3. What are the top 5 highest and lowest salaries?
# 4. Visualise the salaries of professors using a box plot and explain your findings.
# 5. What is the average of years spent during service of a professor and associate professor?
# 6. What is probability that a professor is female and belongs to discipline B?

df = pd.read_csv('./Salaries.csv')
df.head()
```

```python
# 1. Visualise the gender distribution using a pie chart and calculate what percentage of teachers are male.
males = 0
females = 0
for i in range(len(df)):
    if df['sex'][i] == 'Male':
        males += 1
    else:
        females += 1

plt.pie([males, females], labels=["Males", "Females"], autopct='%1.1f%%', shadow=True, startangle=90)
plt.legend()
plt.show()
```

```python
# 3.
salaries = df['salary'].tolist()
salaries.sort()
print(f"Top 5 salaries: {salaries[-5:]}")
print(f"Bottom 5 salaries: {salaries[:5]}")
```

```python
# 4.
B = plt.boxplot(df['salary'])
# Find mean and median of salaries
print([item.get_ydata() for item in B['whiskers']])
# Find median of salaries from the box plot
plt.yticks(color='white')
plt.grid(True)
plt.show()
# FINDINGS :
# The salaries are skewed right.
# The top 4 salaries are actually outliers
```

```python
# 5.
print(f"Average years of service of a Professor: {df[df['rank'] == 'Prof']['yrs.service'].mean()}")
print(f"Average years of service of an Assosiate Professor: {df[df['rank'] == 'AsstProf']['yrs.service'].mean()}")
```
